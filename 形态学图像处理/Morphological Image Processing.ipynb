# 形态学图像处理

```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
import morphology
```

## 腐蚀

```python
_, axe = plt.subplots(1, 4, figsize = (15, 8))

img = cv.imread('./image/Fig0905(a)(wirebond-mask).tif')
axe[0].imshow(img)

kernel = np.ones((11, 11))
img_e = cv.erode(img, kernel)
axe[1].imshow(img_e)

kernel = np.ones((15, 15))
img_e = cv.erode(img, kernel)
axe[2].imshow(img_e)

kernel = np.ones((45, 45))
img_e = cv.erode(img, kernel)
axe[3].imshow(img_e)
```

## 膨胀

```python
_, axe = plt.subplots(1, 2, figsize = (15, 8))

img = cv.imread('./image/Fig0907(a)(text_gaps_1_and_2_pixels).tif')
axe[0].imshow(img)

kernel = np.ones((3, 3))
img_d = cv.dilate(img, kernel)
axe[1].imshow(img_d)
```

## 开运算和闭运算

```python
_, axe = plt.subplots(2, 3, figsize = (15, 8))

img = cv.imread('./image/Fig0911(a)(noisy_fingerprint).tif')
axe[0, 0].imshow(img)

#腐蚀
kernel = np.ones((3, 3))
img_e = cv.erode(img, kernel)
axe[0, 1].imshow(img_e)

#膨胀
img_o = cv.dilate(img_e, kernel)
axe[0, 2].imshow(img_o)

#开运算的膨胀
img_o_d = cv.dilate(img_o, kernel)
axe[1, 0].imshow(img_o_d)

#开运算的闭运算
img_o_c = cv.erode(cv.dilate(img_o, kernel), kernel)
axe[1, 1].imshow(img_o_c)
```

## 边界提取

```python
_, axe = plt.subplots(1, 2, figsize = (15, 8))

img = cv.imread('./image/Fig0914(a)(licoln from penny).tif')
axe[0].imshow(img)

kernel = np.ones((3, 3))
img_b = (img - cv.erode(img, kernel))
axe[1].imshow(img_b)
```

## 孔洞填充

```python
_, axe = plt.subplots(1, 2, figsize = (15, 8))

img = cv.imread('.\image\Fig0916(a)(region-filling-reflections).tif')
axe[0].imshow(img)

kernel = np.ones((9, 9))
img_hf = morphology.hole_fill(img, kernel, (0, 27))
axe[1].imshow(1 - img_hf)
```
