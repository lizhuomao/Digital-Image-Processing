# 彩色图像处理

```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
import colorimageprocess as cip
```

```python
img = cv.imread('./image/Fig0638(a)(lenna_RGB).tif')
b, g, r = cv.split(img)
img_rgb = cv.merge([r, g, b])
plt.imshow(img_rgb)
```

## RGB分量

```python
_, axe = plt.subplots(2, 2, figsize = (15, 8))

axe[0, 0].imshow(img_rgb)
r, b, g = cv.split(img_rgb)
axe[0, 1].imshow(r, cmap = 'gray')
axe[1, 0].imshow(g, cmap = 'gray')
axe[1, 1].imshow(b, cmap = 'gray')
```

```python
_, axe = plt.subplots(1, 3, figsize = (15, 8))
img_hsi = cv.cvtColor(img, cv.COLOR_BGR2HSV)
h, s, i = cv.split(img_hsi)

axe[0].imshow(h, cmap = 'gray')
axe[1].imshow(s, cmap = 'gray')
axe[2].imshow(i, cmap = 'gray')
```

## 图像平滑

```python
_, axe = plt.subplots(1, 3, figsize = (15, 8))

#rgb平滑
img_s = cip.smoothing_RGB(img_rgb, 5, 5)
axe[0].imshow(img_s)

#hsi平滑
img_s_hsi = cip.smoothing_HSI(img_hsi, 5, 5)
img_s_hsi = cv.cvtColor(img_s_hsi, cv.COLOR_HSV2BGR)
b, g, r = cv.split(img_s_hsi)
img_s_hsi = cv.merge([r, g, b])
axe[1].imshow(img_s_hsi)

#两个结果的差
axe[2].imshow(img_s - img_s_hsi)
```

## 图像锐化

```python
#结果非常不理想 。。。。
#换了边缘核也不行
_, axe = plt.subplots(1, 3, figsize = (15, 8))

#rgb平滑
img_s = cip.sharpening_RGB(img_rgb)
axe[0].imshow(img_rgb + img_s)

#hsi平滑
img_s_hsi = cip.sharpening_HSI(img_hsi)
img_s_hsi = cv.cvtColor(img_s_hsi, cv.COLOR_HSV2BGR)
b, g, r = cv.split(img_s_hsi)
img_s_hsi = cv.merge([r, g, b])
axe[1].imshow(img_rgb + img_s_hsi)

#两个结果的差
axe[2].imshow(img_s - img_s_hsi)
```
