# 图片压缩和水印

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
import imagecompression as ic
```

## 比特平面编码

```python
img = cv.imread('./image/Fig0819(a).tif', cv.IMREAD_GRAYSCALE)
img_g = ic.graycode(img, 8)
plt.imshow(img, cmap = 'gray')
```

```python
_, axe = plt.subplots(4, 2, figsize = (15, 15))

axe[0, 0].imshow(img, cmap = 'gray')
axe[0, 1].imshow(ic.getbiteplane(img, 7), cmap = 'gray')

axe[1, 0].imshow(ic.getbiteplane(img, 6), cmap = 'gray')
axe[1, 1].imshow(ic.getbiteplane(img_g, 6), cmap = 'gray')

axe[2, 0].imshow(ic.getbiteplane(img, 5), cmap = 'gray')
axe[2, 1].imshow(ic.getbiteplane(img_g, 5), cmap = 'gray')

axe[3, 0].imshow(ic.getbiteplane(img, 4), cmap = 'gray')
axe[3, 1].imshow(ic.getbiteplane(img_g, 4), cmap = 'gray')
```
